```python
#importing necessary libraries
import numpy as np
import random
import itertools
import math
```

```python
pop_size = 6 
#intializing the vector that is going to hold the probabilities of parents being selected with zeros
Ps = [0]*pop_size
# variable to store total population
totpop = []
gen = 3
```

```python
class GA:
    def generate_population(self,pop_size):
        # each row of the population matrix is the chromosome of the individual
        # intializing population matrix of size pop_size x pop_size with zeros
        population = np.zeros((pop_size,pop_size))
        for i in range(pop_size):
            # random.sample generates unique random numbers and sets it to the population matrix
            population[i] = random.sample(range(0,100),pop_size)
        #random.sample generates float values so converting it into int type
        population = population.astype(int)
        print("INITIALLY : ")
        print()
        return population 

    def FitnessCalc(self,population):
        global Ps
        print("POPULATION : ")
        #printing the population at hand
        print(population)
        print()
        #defining the fitness function vector as row sums of the population matrix, this will be the total distances vector 
        fitnessfn = np.sum(population, axis=1)
        print("FITNESS OF POPULATION : ")
        print()
        print(fitnessfn)
        ffj = 0
        # the probability of the individual being selected or probability of the contribution of the individual 
        # in the population which is calculated by f[i]/sum of all fitnesses from o to pop_size
        for j in range(0,pop_size):
            ffj = ffj + fitnessfn[j]
        for i in range(0,pop_size):
            Ps[i] = fitnessfn[i]/ffj #Ps is the probabilities of getting selected as parents  
        self.selection(population,fitnessfn)
    
    def selection(self,population,fitnessfn):
        global Ps
        temp = Ps.copy()
        tempp = [0]*pop_size
        pscopy = list(Ps.copy())
        pscopy.sort()
        # we are taking pop_size/2 number of individuals whose distances are minimmum
        # we are going by the logic minimum distance maximum fitness
        for i in range(pop_size):
            for j in range(pop_size):
                if(pscopy[i]==temp[j]):
                    tempp[i] = tempp[i] + population[j] 
        # we are deleting the rest Pop_size/2 amount of individuals with maximum distances
        del tempp[math.ceil(pop_size/2):pop_size]
        # we are creating a list of tuples consisting of all the combinations that can be made 
        # using the pop_size/2 individuals
        parents = itertools.combinations(tempp,2)
        p12 = list(parents)
        # We iterate through the above list and take one combination at a time and call the crossover function
        for i in range(len(p12)):
            self.singleptcrossover(p12[i][0],p12[i][1],population)
    
    def singleptcrossover(self,parentone,parenttwo,population):
        childone = []
        childtwo = []
        new_population = []
        # we are generating a random index to select the point/index of crossover
        c1 = np.random.randint(1,len(parentone)-1)
        # crossing over and getting the two children
        childone = np.concatenate((parentone[0:c1], parenttwo[c1:]), axis=None)
        childtwo = np.concatenate((parenttwo[0:c1], parentone[c1:]), axis=None)
        #calling the mmutate function
        self.mutate(childone,childtwo)

    def mutate(self,childone,childtwo):
        # here the algorithm using swap mutation
        temp1 = []
        temp2 = []
        # we are generating two indices whose elements we are gonna swap for the mutation process
        m1 = np.random.randint(1,len(childone))
        m2 = np.random.randint(1,len(childtwo))
        # mutating the two children chromosomes
        while(m1 == m2):
            m2 = np.random.randint(1,len(childtwo))

        temp1 = childone[m1]
        childone[m1] = childone[m2]
        childone[m2] = temp1

        m1 = np.random.randint(0,len(childone))
        m2 = np.random.randint(0,len(childtwo))
        while(m1 == m2):
            m2 = np.random.randint(1,len(childtwo))
        temp2 = childtwo[m1]
        childtwo[m1] = childtwo[m2]
        childtwo[m2] = temp2 
        # we are apprending the children to the totpop as they will act as the population for the next generation
        totpop.append(childone)
        totpop.append(childtwo)    
```

```python
global Ps,totpop
tsp = GA()
#initial population generation 
p = tsp.generate_population(pop_size)
#fitness calculation of the initial population
tsp.FitnessCalc(p)
# converting the population matrix into a 2d array
p = np.asarray(totpop)
# if the pop_size is an odd number then the new population will have +1 extra individual as while calculating calculation
# if Pop_size/2 population, we are rounding off the pop_size (which is an odd number to the highest near value)
# what we are doing is we are going to be taking the individual with the least fitness here, that is nothing but highest
# or maxinmum distance and we are removing that individual from the population
if((pop_size%2)!=0):
    fn = np.sum(p, axis=1)
    for k in range(len(fn)):
        if(fn[k]==max(fn)):
            ind = k
    newp = np.delete(p,ind,axis=0)
    p = newp
# this part is similar as the top but here the difference is you are doing for multiple generations, with the initial
# population's children population is  being fed in first as the first input and further on, the children population of the 
# the previous generation is fed as input
# the process for the pop_size being odd is the same as above that is being done here
Ps = [0]*pop_size
totpop = []
print("==================================")
for i in range(gen):
    print("GENERATION : ",i+1)
    print()
    tsp.FitnessCalc(p)
    print("==================================")
    p = np.asarray(totpop)
    if((pop_size%2)!=0):
        fn = np.sum(p, axis=1)
        for k in range(len(fn)):
            if(fn[k]==max(fn)):
                ind = k
        newp = np.delete(p,ind,axis=0)
        p = newp
    Ps = [0]*pop_size
    totpop = []
```
